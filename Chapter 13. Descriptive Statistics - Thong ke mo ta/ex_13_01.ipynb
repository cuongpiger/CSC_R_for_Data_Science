```r
options(repr.plot.width=13, repr.plot.height=10)
ggplot(stack(pima[, 1:8]), aes(x = ind, y = values)) + 
    geom_boxplot() + xlab("") + ylab("Value") + theme_bw(base_size = 14) +
    theme(axis.text.x = element_text(face = 'bold', color = 'red', size = 15),
          axis.text.y = element_text(face = 'bold', color = 'blue', size = 15))
```

```r
df <- data.frame(Feature=character(),
                 Quantile=numeric(), 
                 Value=numeric(), 
                 stringsAsFactors=FALSE) 
quantile = c(0.05, 0.15, 0.25, 0.50, 0.75, 0.90)

for (cn in colnames(pima)[1:8]) {
    feature = rep_len(cn, length(quantile))
    value = as.vector(quantile(as.vector(pima[, cn]), quantile))
    tmp_df = data.frame(Feature=feature, Quantile=quantile, Value=value, stringsAsFactors=FALSE)
    df = bind_rows(df, tmp_df)
}
```

```r
ggplot(df, aes(x = Quantile, y = Value, fill = Feature)) +
    geom_line() +
    geom_point(size = 4, shape = 21)
```
