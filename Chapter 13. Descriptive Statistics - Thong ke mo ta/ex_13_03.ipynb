```r
library('xlsx')
library('e1071')
```

```r
df = read.xlsx("./data/students.xls", sheetName = 'Full')
```

```r
head(df)
```

```r
df = df[colSums(!is.na(df)) > 0]
df = na.omit(df)
```

```r
df = df[, 4:14]
```

```r
summary(df)
```

```r
major = table(df$Major)
colors1 = c('red', 'yellow', 'green')

pie(major, col=colors1, main='Number of Students per Major')
```

```r
gender = table(df$Gender)
colors2 = c('blue', 'pink')

pie(gender, col=colors2, main="Male vs Female")
```

```r
status = table(df$Student.Status)
colors3 = c('violet', 'cyan')

pie(status, col=colors3, main='Graduate vs Undergraduate')
```

```r
sapply(df[7:11], var)
```

```r
sapply(df[7:11], sd)
```

```r
skew = apply(df[,7:11], 2, skewness)
skew
```

```r
kur = apply(df[,7:11], 2, kurtosis)
kur
```

```r
par(mfrow=c(2, 2))
hist(df$Age, main='Ages', xlab='Ages', xlim=c(min(df$Age) - 5, max(df$Age) + 5),
     col='orange', breaks=10, border='blue', las=1, freq=FALSE)
lines(density(df$Age))

hist(df$SAT, main='SAT', xlab='SAT', xlim=c(min(df$SAT) - 10, max(df$SAT) + 10),
     col='orange', breaks=10, border='blue', las=1, freq=FALSE)
lines(density(df$SAT))

hist(df$Average.score..grade., main='Avg Score Grade', xlab='Ages',
     xlim=c(min(df$Average.score..grade.) - 10, max(df$Average.score..grade.) + 10),
     col='orange', breaks=10, border='blue', las=1, freq=FALSE)
lines(density(df$Average.score..grade.))

hist(df$Height..in., main='Height in', xlab='Height in',
     xlim=c(min(df$Height..in.) - 5, max(df$Height..in.) + 5),
     col='orange', breaks=10, border='blue', las=1, freq=FALSE)
lines(density(df$Height..in.))
```
