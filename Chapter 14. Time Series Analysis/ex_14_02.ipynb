```r
births = scan("./data/mybirths.dat.txt", skip=3)
```

```r
births
```

```r
births_ts = ts(births, start = c(1946, 1), frequency = 12)
```

```r
births_ts
```

```r
plot(births_ts, main = "Births per month in New Your city\nfrom 01/1946 to 12/1959",
     col = 'red', xlab = 'Year', ylab = 'Births')
```

> **Nhận xét**
> * Từ năm 1950, số trẻ sinh ra có xu hướng tăng theo năm.
> * Tập dữ liệu này cho thấy dữ liệu mang tính cyclic và seasonal.

* Xem các thành phần trong dữ liệu

```r
comp = decompose(births_ts)
```

```r
comp$x
```

```r
comp$seasonal
```

> **Nhận xét**
> * Xét yếu tố theo mùa (seasonal) thì tháng 1, 3, 8, 9, 10, 11 có xu hướng giảm, còn lại tăng.

```r
comp$trend
```

> **Nhận xét**
> * Số lượng sinh có xu hướng tăng dần từ năm 1950

```r
comp$random
```

> **Nhận xét**
> * Giảm rồi tăng nhưng ko đồng đều, xem ra ko có tính theo chu kì

```r
plot(comp)
```

> * Do `trend` là phi tuyến nên dùng `multiplicative`

```r
births.timeseries.forecasts = HoltWinters(births_ts, seasonal = 'multiplicative')
```

```r
births.timeseries.forecasts
```

> **Nhận xét**:
> * $alpha$ bằng 0.5, cho thấy giá trị hiện tại phụ thuộc vào dữ liệu gần đây và kể cả trong quá khứ
> * $beta$ bằng 0 tức độ dốc ko cập nhật theo thời gian qua.
> * $gamma$ bằng 0.6 ảnh hưởng của yếu tố mùa tại thời điểm hiện tại phụ thuộc vào thời gian gần.

```r
births.timeseries.forecasts$SSE
```

```r
births.timeseries.forecasts$SSE / length(births_ts)
```
