# 2. Hold-Winters
## 2.1. Exponential Smoothing
### 2.1.2. Simple Exponential Smoothing (SES)

```r
ts_rainfall = ts(rain_fall, frequency = 12, start = c(2014, 1))

# new
forecast_rainfall = HoltWinters(ts_rainfall, beta = FALSE, gamma = FALSE)
```

```r
forecast_rainfall
```

```r
forecast_rainfall$SSE
```

```r
plot(forecast_rainfall)
```

* Tính $MSE$

```r
forecast_rainfall$SSE / length(ts_rainfall)
```

* Tính $RMSE$

```r
sqrt(forecast_rainfall$SSE / length(ts_rainfall))
```

* Cài thư viện để dự đoán tương lai

```r
install.packages('forecast')
```

```r
library(forecast)
```

```r
forecast_rainfall_new = forecast:::forecast.HoltWinters(forecast_rainfall, h=5) # 5 tháng tiếp theo
```

```r
forecast_rainfall_new
```

```r
forecast:::plot.forecast(forecast_rainfall_new)
```

```r
# cách 2
plot(forecast_rainfall_new)
```

> **Nhận xét**
> * SSE ko phù hợp cho dataset này.

### 2.1.2. Holt Exponential Smoothing (HES)

```r
HES.forecast = HoltWinters(log(ts_rainfall), gamma = FALSE)
```

```r
HES.forecast
```

```r
HES.forecast$SSE
```

* Tính MSE

```r
HES.forecast$SSE / length(rain_fall)
```

* Tính RMSE

```r
sqrt(HES.forecast$SSE / length(rain_fall))
```

```r
HES.forecast.new = forecast:::forecast.HoltWinters(HES.forecast, h = 5)
```

```r
HES.forecast.new
```

```r
plot(HES.forecast.new)
```

* Un-log

```r
y = data.frame(HES.forecast.new)
y$Forecast = exp(y$Point.Forecast)

y
```

### 2.1.3. Holt-Winters Exponential Smoothing (HWSE)

```r
HWSE.forecast = HoltWinters(log(ts_rainfall))
```

```r
HWSE.forecast
```

```r
plot(HWSE.forecast)
```

```r
HWSE.forecast$SSE
```

```r
HWSE.forecast$SSE / length(rain_fall)
```

```r
sqrt(HWSE.forecast$SSE / length(rain_fall))
```

```r
HWSE.forecast.new = forecast:::forecast.HoltWinters(HWSE.forecast, h = 5)
```

```r
HWSE.forecast.new
```

```r
plot(HWSE.forecast.new)
```

```r
x = data.frame(HWSE.forecast.new)
x$Forecast = exp(x$Point.Forecast)

x
```
