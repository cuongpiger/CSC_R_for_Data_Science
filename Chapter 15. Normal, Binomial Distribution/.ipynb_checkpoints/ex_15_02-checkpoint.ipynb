```r
heights = dnorm(seq(140, 180, by = .5), mean = 156, sd = 4.6)
```

```r
print(heights)
```

```r
plot(seq(140, 180, by = .5), heights,
     ylab = "f(height)", xlab = 'Height',
     main = 'Probability distribution of height in Vietnamese women')
```
