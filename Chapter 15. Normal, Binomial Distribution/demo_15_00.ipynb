# 1. Normal Distribution
## 1.1. `dnorm()`

```r
x = seq(-10, 10, by = .1)
y = dnorm(x, mean = 0, sd = 1)

plot(x, y)
```

* Nếu $\mu = 156cm, \sigma = 4.6cm \text{, nữ cao} = 160cm$ thì distribution values là bao nhiêu 

```r
dnorm(160, mean = 156, sd = 4.6)
```

## 1.2. `pnorm`

```r
x = seq(-10, 10, by = .1)
y = pnorm(x, mean = 0, sd = 1)

plot(x, y)
```

* Nếu $x = 5$ thì $p$ bằng bao nhiêu

```r
pnorm(5, mean = 0, sd = 1)
```

## 1.3. `qnorm()`

```r
x = seq(0, 1, 0.02)
y = qnorm(x, mean = 0, sd = 1)

plot(x, y)
```

```r
qnorm(0.8, mean = 0, sd = 1)
```

## 1.4. `rnorm()`

```r
y = rnorm(50)

hist(y, main = "Normal distribution")
```

```r
y = rnorm(50, mean = 156, sd = 4)

hist(y, main = "Normal distribution")
```

# 2. Binomial Distribution
## 2.1. `dbinom()`

```r
x = seq(0, 50, by = 1)
y = dbinom(x, size = 50, prob = 0.5)

plot(x, y)
```

```r
dbinom(17, size = 50, prob = 0.5)
```

* Có 10 học sinh, trong đó có 6 nữ, nếu 3 bạn nữ dc chọn ngẫu nhiên thì xác suất để có 2 bạn nữ là bao nhiêu?

  * $p(\text{dc 1 nữ}) = 0.6, k = 2, n = 3$

```r
dbinom(2, 3, 0.6)
```

## 2.2. `pbinom`, xác suất tích lũy

* Xác suất nhận dc $\leq$ 25 mặt head sau 50 lần tung một đồng xu

```r
pbinom(25, 50, 0.5)
```

* Xác suất nhận dc $\geq$ 25 mặt head sau 50 lần tung

```r
1 - pbinom(24, 50, 0.5)
```

## 2.3. `qbinom(x, size, prob)`

* Có bao nhiêu mặt head nhận dc nếu xác suất nhận dc là 0.25 sau 50 lần tung một đồng xu

```r
qbinom(.25, 50, 0.5)
```

## 2.4. `rbinom(n, size, prob)`

* Tìm 8 giá trị ngẫu nhiên từ 100 mẫu với xác suất là $\frac{1}{6}$

```r
rbinom(8, 100, 1/6)
```
