```r
library(ggplot2)
library('gridExtra')
library('ggExtra')
library(tidyverse)
```

```r
economy = read.csv("./data//EconomistData.csv")
```

```r
head(economy)
```

```r
summary(economy)
```

```r
str(economy)
```

```r
options(repr.plot.width=10, repr.plot.height=8)
plot = ggplot(data = economy, aes(x = CPI,
                                  y = HDI,
                                  fill = Region,
                                  color = Region,
                                  lwd = 1))
plot = plot + geom_point(aes(size = HDI.Rank))
plot = plot + ylab("HDI")
plot = plot + xlab('CPI')
plot = plot + ggtitle("CPI, HDI and HDI.Rank follow Region")
plot = plot + theme_bw(base_size = 14)
plot = plot + theme(legend.position = 'bottom',
                    plot.title = element_text(color = 'red', size = 20, face = 'bold', hjust = 0.5),
                    axis.text.x = element_text(face = 'bold', color = 'red'),
                    axis.text.y = element_text(face = 'bold', color = 'blue'))
plot
```

```r
plot = ggplot(data = economy, aes(x = CPI, y = HDI)) + 
       geom_point() + geom_smooth(method = 'lm') + # vẽ đường phi tuyến
       ggtitle("CPI vs HDI") + 
       theme_bw(base_size = 14) + 
       theme(plot.title = element_text(color = 'red', size = 20,
                                       face = 'bold', hjust = 0.5)) 

plot
```
