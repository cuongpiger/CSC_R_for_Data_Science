```r
library('ggplot2')

plot = ggplot(chol, aes(AGE)) + 
     geom_histogram(color = 'white', fill = 'blue') +
     ggtitle("Histogram on AGE feature") + 
     theme_bw(base_size = 14) +
     theme(plot.title = element_text(color = 'red', size = 20,
                                     face = 'bold', hjust = 0.5))

plot
```

```r
plot = ggplot(data = chol, aes(x = WEIGHT, y = HEIGHT)) + 
       geom_point() + geom_smooth(method = 'lm') + # vẽ đường phi tuyến
       ggtitle("WEIGHT vs HEIGHT") + 
       theme_bw(base_size = 14) + 
       theme(plot.title = element_text(color = 'red', size = 20,
                                       face = 'bold', hjust = 0.5)) 

plot
```

```r
plot = ggplot(data = chol, aes(x = WEIGHT, y = HEIGHT))
plot = plot + geom_point()
plot = plot + scale_x_continuous(breaks = seq(50, 120, 30), limits = range(50, 120))
plot = plot + scale_y_continuous(breaks = seq(150, 200, 30), limits = range(150, 200))
plot = plot + xlab('Weight')
plot = plot + ylab("Height")
plot = plot + ggtitle("WEIGHT vs HEIGHT")
plot = plot + theme_bw(base_size = 14)
plot = plot + theme(plot.title = element_text(color = 'red', size = 16, face = 'bold', hjust = 0.5),
                    axis.text.x = element_text(face = 'bold', color = 'red', angle = 45),
                    axis.text.y = element_text(face = 'bold', color = 'blue'))

# new
ggMarginal(plot, type = 'histogram', col = 'white', fill = 'blue', bins = 20)
```

```r
chol = chol %>% mutate(WEIGHT.bin = ifelse(WEIGHT <= 70, "<= 70", 
                                    ifelse(WEIGHT <= 90, "(70, 90]",
                                    ifelse(WEIGHT <= 110, "(90, 110]", ">= 110"))))
```

```r
head(chol)
```

```r
plot = ggplot(data = chol, aes(x = WEIGHT.bin, y = HEIGHT, fill = WEIGHT.bin)) +
       ggtitle("HEIGHT based on binned WEIGHT") +
       xlab('WEIGHT') +
       geom_boxplot() + geom_jitter(alpha = 0.1) + theme_bw(base_size = 14) +
       theme(plot.title = element_text(color = 'red', size = 20, face = 'bold', hjust = 0.5),
             axis.text.x = element_text(face = 'bold', color = 'red'),
             axis.text.y = element_text(face = 'bold', color = 'blue')) +
       guides(fill=guide_legend(title="WEIGHT"))

plot
```

```r
pie_data = chol %>% group_by(WEIGHT.bin) %>% count() %>% 
           ungroup() %>% mutate(per = `n`/sum(`n`)) %>%
           arrange(desc(WEIGHT.bin))
pie_data$label = scales::percent(pie_data$per)
```

```r
pie_data
```

```r
pie_weight = ggplot(data=pie_data)+
                    geom_bar(aes(x="", y=per, fill=WEIGHT.bin), stat="identity", width = 1)+
                    coord_polar("y", start=0)+
                    theme_void()+
                    geom_text(aes(x=1, y = cumsum(per) - per/2, label=label)) +
             ggtitle("WEIGHT") +
             theme(plot.title = element_text(color = 'red', size = 20, face = 'bold', hjust = 0.5))

pie_weight
```

```r
chol = chol %>% mutate(HEIGHT.bin = ifelse(HEIGHT <= 160, "<= 160", 
                                    ifelse(HEIGHT <= 170, "(160, 170]",
                                    ifelse(HEIGHT <= 180, "(170, 180]", '>= 180'))))
```

```r
pie_data = chol %>% group_by(HEIGHT.bin) %>% count() %>% 
           ungroup() %>% mutate(per = `n`/sum(`n`)) %>%
           arrange(desc(HEIGHT.bin))
pie_data$label = scales::percent(pie_data$per)
```

```r
pie_data
```

```r
pie_height = ggplot(data=pie_data)+
                    geom_bar(aes(x="", y=per, fill=HEIGHT.bin), stat="identity", width = 1)+
                    coord_polar("y", start=0)+
                    theme_void()+
                    geom_text(aes(x=1, y = cumsum(per) - per/2, label=label)) +
             ggtitle("HEIGHT") +
             theme(plot.title = element_text(color = 'red', size = 20, face = 'bold', hjust = 0.5))

pie_height
```

```r
grid.arrange(pie_weight, pie_height, nrow = 2)
```
