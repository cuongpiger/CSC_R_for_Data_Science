#### 1. Đọc dữ liệu, hiển thị thông tin chung của dữ liệu : head(), tail(), str(), summary()

```r
str(ames_housing)
```

```r
summary(ames_housing)
```

#### 2. Cho biết số dòng, số cột của dữ liệu

```r
ames_housing_dim = dim(ames_housing)
writeLines(sprintf("Số dòng %d, số cột %d", 
                   ames_housing_dim[1],
                   ames_housing_dim[2]))
```

#### 3. Cho biết có bao nhiêu loại Garage.Type, đó là những loại nào, mỗi loại đếm được bao nhiêu mẫu. 

```r
table(ames_housing$Garage.Type)
```

#### 4. Xóa các cột Garage.Qual, Garage.Cond, Pool.QC, Fence, Misc.Feature trong dữ liệu

```r
ames_housing$Garage.Qual = NULL
ames_housing$Garage.Cond = NULL
ames_housing$Pool.QC = NULL
ames_housing$Fence = NULL
ames_housing$Misc.Feature = NULL
```

```r
dim(ames_housing)
```

#### 5. Tìm max, min của Gr.Liv.Area theo Electrical và Heating (sử dụng group_by() và summarize()). 

```r
library("tidyverse")
```

```r
ames_housing %>% group_by(Electrical, Heating) %>% summarise(
    min.Gr.Liv.Area=min(Gr.Liv.Area, na.rm=TRUE),
    max.Gr.Liv.Area=max(Gr.Liv.Area, na.rm=TRUE), .groups = 'drop')
```

#### 6. Vẽ biểu đồ thể hiện mối liên hệ của SalePrice và X1st.Flr.SF. Nhận xét biểu đồ. 

```r
sale_price = ames_housing$SalePrice
x1st = ames_housing$X1st.Flr.SF
```

```r
plot(sale_price, x1st,
     main = "SalePrice vs X1st.Flr.SF",
     xlab = "SalePrice", ylab = "X1st.Flr.SF",
     pch = 19, frame = FALSE, col = 'blue')
lines(lowess(sale_price, x1st), col = "red", lwd = 3)
```

#### 7. Vẽ pie chart thể hiện % giữa Y(1) và N(0) của cột Central.Air

```r
central_air = ames_housing$Central.Air
central_air_freq = table(central_air)
central_air_names = names(central_air_freq)
piepercent = round(100 * central_air_freq/sum(central_air_freq), 1)
```

```r
pie(central_air_freq, labels = paste(piepercent, "%"), 
    main = "Central Air",
    col = rainbow(length(central_air_freq)),
    clockwise = TRUE)
legend("topright", central_air_names, cex = 0.9,
       fill = rainbow(length(central_air_freq)), y.intersp = 1)
```

#### 8. Cho biết năm xây của các căn nhà cũ nhất và mới nhất (theo Year.Built). Liệt kê các căn nhà cũ nhất, mới nhất với 3 thông tin Id, Year.Built, SalePrice

```r
ames_housing_oldest_value = min(ames_housing$Year.Built)
ames_housing_oldest_lists = ames_housing %>% filter(
    Year.Built == ames_housing_oldest_value) %>% select(Id, Year.Built, SalePrice)

ames_housing_oldest_lists
```

```r
ames_housing_newest_value = max(ames_housing$Year.Built)
ames_housing_newest_lists = ames_housing %>% filter(
    Year.Built == ames_housing_newest_value) %>% select(Id, Year.Built, SalePrice)

ames_housing_newest_lists
```

#### 9. Thống kê số lượng các căn nhà được xây theo từng năm. In head và tail.  Cho biết năm nào có nhiều nhà được xây nhất? 

```r
built_houses = ames_housing %>% group_by(Year.Built) %>% summarise(Year.Built.Count=n())
```

```r
head(built_houses)
```

```r
tail(built_houses)
```

```r
t(built_houses)
```

```r
id_max_year_built_freq = which.max(built_houses$Year.Built.Count)

id_max_year_built_freq
```

```r
writeLines(sprintf("Năm có nhiều nhà xây nhất là năm %d với %d căn nhà được xây",
                   built_houses$Year.Built[id_max_year_built_freq],
                   built_houses$Year.Built.Count[id_max_year_built_freq]))
```

#### 10. Trực quan hóa kết quả của câu thống kê trên với 10 năm gần đây nhất bằng barplot; với tất cả các năm bằng line.

```r
ten_closest_houses = tail(built_houses, 10)
```

```r
ten_closest_houses
```

```r
plot(ten_closest_houses$Year.Built.Count, type = 'o',
     col = 'red', xlab = 'Year', ylab = 'Numbers',
     main = 'Number of houses built year by year', xaxt = 'n', frame = FALSE)
axis(1, at = 1:dim(ten_closest_houses)[1], labels = ten_closest_houses$Year.Built)
```

#### 11. Vẽ boxplot của cột SalePrice. 

```r
boxplot(ames_housing$SalePrice, main = "SalePrice", ylab = "Price", col = 'blue')
```

#### 4. Vẽ boxplot cho knowledge_score và cho calcium_intake. knowledge_score có outlier hay không? calcium_intake có outlier hay không? Nhận xét

```r
boxplot(canxi$knowledge_score, main = "Knowledge Score", ylab = "Score", col = 'blue')
```

```r
boxplot(canxi$calcium_intake, main = "Calcium Intake", ylab = "Calcium", col = 'yellow')
```

#### 5. Vẽ biểu đồ thể hiện mối quan hệ giữa knowledge_score và calcium_intake. Nhận xét.

```r
knowledge_score = canxi$knowledge_score
calcium_intake = canxi$calcium_intake
```

```r
plot(knowledge_score, calcium_intake,
     main = "Knowledge vs Canxi",
     xlab = "Knowledge", ylab = "Canxi",
     pch = 19, frame = FALSE, col = 'green')
lines(lowess(knowledge_score, calcium_intake), col = "red", lwd = 3)
```

> **Nhận xét**
> * Có mối quan hệ tuyến tính giữa **Canxi** và **Knowledge**.
