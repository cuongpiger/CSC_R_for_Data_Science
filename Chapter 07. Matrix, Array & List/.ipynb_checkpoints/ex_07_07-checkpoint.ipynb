```r
year = 2008:2018
month = 1:12
day = 1:31

ymd = list('year'=year, 'month'=month, 'day'=day)
```

```r
ymd
```

```r
ymd$year = 2010:2020

ymd
```

```r
ymd$year = ymd$year[-1]

ymd
```

```r
x = c(1, 3, 4, 7, 11, 18, 29)
cal_list = list("x*2"=x*2, "x/2"=x/2, "sqrt(x)"=sqrt(x))
```

```r
cal_list
```

```r
cal_list$`sqrt(x)`[3:5]
```

```r
ymd.cal_list = c(ymd, cal_list)

ymd.cal_list
```

```r
year_x2_day = ymd.cal_list[c('year', 'x*2', 'day')]

year_x2_day
```

```r
length(year_x2_day$`x*2`)
```

```r
tail(year_x2_day$year, 1)[1]
```
