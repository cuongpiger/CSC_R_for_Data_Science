```r
year = 2014:2017
first_fallen = c(309, 271, 263, 255)
one_hundred_fallen = c(309, 314, 312, 304)
```

```r
data = matrix(c(first_fallen, one_hundred_fallen), nrow = 2, byrow = TRUE)
```

```r
data
```

```r
barplot(data, main = "First and last dates of leaf fall in Yellow Birch",
        xlab = "Year", ylab = "Numbers", beside = TRUE,
        col = c("brown","blue"), names.arg = year)
legend("bottomleft", c('First fallen', "100% fall"),
       cex = 0.9, fill = c('brown', 'blue'), y.intersp = 2)
```
