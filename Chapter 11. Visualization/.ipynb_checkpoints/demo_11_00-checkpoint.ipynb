```r
x = c(46, 37, 35, 53)
labels = c("10A1", '10A2', '10A3', "10A4")
pie(x, labels, main="Pie chart demo", col=rainbow(length(x)), clockwise=TRUE)
```

## 2.2. Theo phần trăm

```r
x = c(46, 37, 35, 53)
labels = c("10A1", '10A2', '10A3', "10A4")
piepercent = round(100 * x/sum(x), 1)

pie(x, labels = paste(piepercent, "%"), 
    main = "Pie chart demo",
   col = rainbow(length(x)),
   clockwise = TRUE)
legend("topright", labels, cex = 0.8, fill = rainbow(length(x)))
```

## 2.3. 3D pie chart

```r
library("plotrix")

x = c(46, 37, 35, 53)
labels = c("10A1", '10A2', '10A3', "10A4")

pie3D(x, labels = labels, main = "Pie chart demo", col = rainbow(length(x)))
```

## 3.2. Xếp chồng lên nhau
* Dưới đây minh họa các vị trí của `legend()`
  ![](images/11_00.png)

```r
colors = c("green", "orange", "brown", "red")
classes = c("10A1", "10A2", "10A3")
grades = c("Excellent", "Very Good", "Good", "Medium")
values = matrix(c(5, 6, 7, 15, 14, 13, 20, 21, 22, 5, 4, 3),
               nrow = 4, ncol = 3, byrow = TRUE)


barplot(values, main = "Grade of 3 classes", names.arg = classes,
       xlab = "Class", ylab = "Students", col = colors)
legend("right", grades, cex = 0.9, fill = colors, y.intersp = 2)
```

## 6.3. Multiple line trên line graph

```r
bears = c(8, 54, 93, 116, 137)
dolphins = c(150, 77, 32, 11, 6)
whales = c(80, 54, 100, 76, 93)

g_range = range(0, bears, dolphins, whales) # trả về một vector gồm 2 phần tử trong đó
                            # phần tử đầu tiên là min của các thành phần truyền vào
                            # phần tử thứ hai là max của các thành phần truyền vào
plot(bears, type = 'o', col = 'red', xlab = 'Year', ylab = 'Numbers',
    main = "Demo multiple line graph", xaxt = 'n', ylim = g_range)
axis(1, at = 1:5, labels = 2017:2021)
lines(dolphins, type = 'o', col = 'blue')
lines(whales, type = "o", col = 'purple')
legend("top", inset = 0.01, c('Bears', 'Dolphins', 'Whales'), cex = 0.8,
      fill = c("red", 'blue', 'purple'))
```

# 8. Wordcloud
## 8.1. Install dependencies packages

```r
install.packages("tm")  # for text mining
install.packages("SnowballC") # for text stemming
install.packages("wordcloud") # word-cloud generator 
install.packages("RColorBrewer") # color palettes
```

## 8.2. Load thư viện

```r
library("tm")
library("SnowballC")
library("wordcloud")
library("RColorBrewer")
```

## 8.3. Demo

```r
text = readLines("./data/introduction_to_DS_MK.txt")
text
```

```r
docs = Corpus(VectorSource(text))
```

```r
inspect(docs)
```

```r
toSpace = content_transformer(function(x, pattern) gsub(pattern, " ", x))
docs = tm_map(docs, toSpace, "/")
docs = tm_map(docs, toSpace, "@")
docs = tm_map(docs, toSpace, "\\|")
```

```r
inspect(docs)
```

```r
stopwords('english')
```

```r
docs = tm_map(docs, content_transformer(tolower))
docs = tm_map(docs, removeNumbers)
docs = tm_map(docs, removeWords, stopwords('english'))
docs = tm_map(docs, removeWords, c('then', 'just', 'more', ''))
docs = tm_map(docs, removePunctuation)
docs = tm_map(docs, stripWhitespace)
```

```r
dtm = TermDocumentMatrix(docs)
m = as.matrix(dtm)
v = sort(rowSums(m), decreasing = TRUE)
d = data.frame(word = names(v), freq = v)
```

```r
head(d, 10)
```

```r
set.seed(1234)
wordcloud(words = d$word, freq = d$freq, min.freq = 1,
         max.words = 200, random.order = FALSE, rot.per = 0.35,
         colors = brewer.pal(3, "Dark2"))
```

### 8.3.1. Khám phá các từ có tần suất cao và mối quan hệ
#### 8.3.1.1. Xuất 3 từ xuất hiện nhiều nhất trong văn bản

```r
findFreqTerms(dtm, lowfreq = 3)
```

#### 8.3.1.2. Tìm các từ nào xuất hiện cùng từ `data` mà có số lần xuất hiện trên 50%

```r
findAssocs(dtm, terms = "data", corlimit = 0.5)
```

### 8.3.2. Vẽ barplot cho 10 từ xuất hiện nhiều nhất

```r
barplot(d[1:10,'freq'], las = 2, names.arg = d[1:10, 'word'],
       col = 'lightblue', main = "Most frequent words",
       ylab = "Word frequencies")
```
