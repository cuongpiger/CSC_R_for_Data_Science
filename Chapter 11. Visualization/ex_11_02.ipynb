```r
library("plotrix")
```

```r
tree_species = c("Beech", "Chestnut", "Hawthom", "Red maple",
                 "Witch Hazel", "Yellow Birch")
no_trees = c(2, 2, 1, 4, 1, 1)
piepercent = round(100 * no_trees/sum(no_trees), 1)
```

```r
pie3D(no_trees, labels = paste(piepercent, "%"), main = "Tree",
      col = rainbow(length(piepercent)))
legend("topright", tree_species, cex = 0.8, fill = rainbow(length(tree_species)))
```
